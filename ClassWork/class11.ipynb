WAP that prompts a user to take an input a number print its square and if no input is given than keyboard interupt error should be raised 


```python
try:
    num = int(input("Enter a Number: "))
    if(num):
        print(num**2)
except KeyboardInterrupt as err:
    print(err)    
```
