WAP to calculate an area of a triangle

```python
a = int(input("Enter a 1st Edge of a triangle: "))
b = int(input("Enter a 2nd Edge of a triangle: "))
c = int(input("Enter a 3rd Edge of a triangle: "))

s = (a+b+c)/2
area = 0.0

area = (s*(s-a)*(s-b)*(s-c))**(0.5)

print(f"area of triangle with side {a}, {b},and {c} is {area}")
```

WAP to calculate bill amount to be paid for an item given it's quantity sold, price per item,amount,discount and tax 

```python
print(f"{'*'*10} BILL {'*'*10}")

qs = float(input("Quantity sold: "))
ppi = float(input("Price per item: "))
discount = float(input("Discount (%): "))
tax = float(input("Tax (%): "))

amount = qs * ppi
discount_amt = amount * discount / 100
tax_amt = (amount - discount_amt) * tax / 100
total = amount - discount_amt + tax_amt

print("\n{:<18} {:>8}".format("Quantity Sold:", qs))
print("{:<18} {:>8.2f}".format("Price Per Item:", ppi))
print("{:<18} {:>8.2f}".format("Amount:", amount))
print("{:<18} {:>8.2f}".format("Discount:", discount_amt))
print("{:<18} {:>8.2f}".format("Tax:", tax_amt))
print("{:<18} {:>8.2f}".format("Total Amount:", total))
print('*'*30)
```

WAP to calculate student's result based on 2 examination, 1 sport event and 3 activities conducted weightage of activities is 30%,sports is 20% and Examination is 50%

```python

exam1 = float(input("Enter marks in Exam 1 (out of 100): "))
exam2 = float(input("Enter marks in Exam 2 (out of 100): "))
sports = float(input("Enter marks in Sports (out of 100): "))
act1 = float(input("Enter marks in Activity 1 (out of 100): "))
act2 = float(input("Enter marks in Activity 2 (out of 100): "))
act3 = float(input("Enter marks in Activity 3 (out of 100): "))

exam_avg = (exam1 + exam2) / 2
activity_avg = (act1 + act2 + act3) / 3

exam_weighted = exam_avg * 0.50
sports_weighted = sports * 0.20
activity_weighted = activity_avg * 0.30


final_score = exam_weighted + sports_weighted + activity_weighted


print("\n------ Student Result ------")
print(f"Exams Contribution (50%): {exam_weighted:.2f}")
print(f"Sports Contribution (20%): {sports_weighted:.2f}")
print(f"Activities Contribution (30%): {activity_weighted:.2f}")
print(f"Final Result: {final_score:.2f} / 100")
```
