WAP to check whether a person is eligible to vote

```python
def check_voter(age):
    if age >= 18:
        print("You are eligible to vote")
    else:
        print("Not Eligible")
        
check_voter(5)
```

WAP to recongnize the character entered by a user

```python
def recongnize(character):
    if character.isalpha():
        print("It is a Alphabet")
    elif character.isdigit():
        print("It is a digit")
    elif character.isspace():
        print("It's a Space")
    else:
        print("It's a Special Character")
        
```

WAP to check whether he/she is eligible to vote and if not print how much year is he/she is short

```python
def check_voter(age):
    if age >= 18:
        print("You are eligible to vote")
    else:
        print(f"Not Eligible come after: {18-age} years")
        
check_voter(5)
```

WAP to enter any character if the entered character in lowercase convert it to uppercase and do vice-versa

```python
def conversion(character):
    if character.isupper():
        print(character.lower())
    else:
        print(character.upper())
        
conversion('heLLO')
        
        
```
