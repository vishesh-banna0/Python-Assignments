```python
class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def display(self):
        print(self.name)
        print(self.age)
```

```python
p1 = person("vishy",23)
```

```python
p1.display()
```

```python
class teacher(person):
    def __init__(self, name, age, researchArea, exp):
        super().__init__(name, age)   
        self.researchArea = researchArea
        self.exp = exp
        
    def displayparent(self):
        super().display()             
        print(self.researchArea)
        print(self.exp)
```

```python
t1 = teacher("Sonu", 22, "AI", 2)
t1.displayparent()
```

```python
class base1(object):
    def __init__(self):
        print("b1")
        super(base1, self).__init__()   # calls next class in MRO


class base2(object):
    def __init__(self):
        print("b2")


class derived(base1, base2):
    def __init__(self):
        super(derived, self).__init__()   # triggers base1 â†’ base2
        print("d class")


d = derived()
```
