```python
num = 5
print(f"The factorial of {num} is {fact(num)}")

def fact(x):
    if x == 1:
        return 1
    else:
        return x*fact(x-1)
    
```

```python
def sumation(x):
    if x == 1:
        return 1
    else:
        return x+sumation(x-1)

num = 5
print(f"The factorial of {num} is {sumation(num)}")
    
```

```python
def findmin(arr, n):
    if n == 1:
        return arr[0]
    else:
        return arr[n-1] if arr[n-1] < findmin(arr, n-1) else findmin(arr, n-1)
arr = [52,52,65,45,15,68,15,36,2,354,85,47,214,968,12,25,14,65]
n = len(arr)
print(f"minimum = {findmin(arr, n)}")
```

```python
def fib(n, a, b):
    if n == 0:
        return
    else:
        print(a + b, end=" ")
        fib(n - 1, b, a + b)
num1 = 10
print("Fibonacci series:")
print("1 1", end=" ")
fib(num1 - 2, 1, 1)
```

```python
word = "Hello World"
l = len(word)-1
def reverse(s,n):
    if n<0:
        return
    print(s[n],end="")
    reverse(s,n-1)
    
reverse(word,l)
    
    
    
    
    
```

```python
def power(x,y):
    if y == 0:
        return 1
    
    return x*power(x,y-1)
        
power(5,3)
```
