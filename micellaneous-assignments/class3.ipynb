WAP find whether a given number is armstrong number or not

```python
val = int(input("Enter a Value: "))
true_val = val
arm_val = 0

while val != 0:
    arm_val += (val%10)**3
    val //= 10


if true_val == arm_val:
    print("it's a armstrong number")
else:
    print("not a armstrong number")
    

    
    
```

convert decimal value to binary

```python
dec = int(input("Enter a digit: "))
binary = []

while dec != 0:
    binary.append(dec%2)
    dec //= 2
    
print(binary[::-1])
```
