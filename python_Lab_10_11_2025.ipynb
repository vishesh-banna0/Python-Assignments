Write a program to create and display all combinations of letters, selecting each letter from different key in dictionary

* Sample data: {'1':['a','b'], '2':['c','d']}
* Expected Output:
ac, ab, ad, bd

```python
dict = {'1':['a','b'], '2':['c','d']}
for i in dict['1']:
    for j in dict['2']:
        print(i+j, end=', ')
```

write a program to get top three items in a shop.
* sample data: {'item1': 45.50, 'item2': 35, 'item3': 41.30,'item4':55, 'item5':24}
* Expected output:
   - item4 55
   - item1 45.5
   - item3 41.3

```python
dict = {'item1': 45.50, 'item2': 35, 'item3': 41.30, 'item4': 55, 'item5': 24}
dict = sorted(dict.items(), key=lambda x: x[1], reverse=True)
print(dict[:3])
```

3. Write a Python program to find the specified number of maximum values in a given
dictionary.
* Original Dictionary:
{&#39;a&#39;: 5, &#39;b&#39;: 14, &#39;c&#39;: 32, &#39;d&#39;: 35, &#39;e&#39;: 24, &#39;f&#39;: 100, &#39;g&#39;: 57, &#39;h&#39;: 8, &#39;i&#39;: 100}
  - 1 maximum value(s) in the said dictionary:
[&#39;f&#39;]
  - 2 maximum value(s) in the said dictionary:
[&#39;f&#39;, &#39;i&#39;]
  - 5 maximum value(s) in the said dictionary:
[&#39;f&#39;, &#39;i&#39;, &#39;g&#39;, &#39;d&#39;, &#39;c&#39;]

```python
dict = {'a': 5, 'b': 14, 'c': 32, 'd': 35, 'e': 24, 'f': 100, 'g': 57, 'h': 8, 'i': 100}
for i in range(len(dict)):
    print(i+1, 'maximum value(s) in the said dictionary:')
    print(sorted(dict.items(), key=lambda x: x[1], reverse=True)[:i+1])
```

Write a Python function that evaluates mathematical expressions represented as
dictionaries.
- The expression dictionary should support basic operations (+, -, *, /), variables,
and nested expressions.

```python
def evaluate_expression(expression, variables=None):

    if variables is None:
        variables = {}

    if 'value' in expression:
        return expression['value']
    elif 'variable' in expression:
        return variables.get(expression['variable'], 0)  # Return 0 if variable not found
    elif 'operator' in expression and 'operands' in expression:
        operator = expression['operator']
        operands = [evaluate_expression(operand, variables) for operand in expression['operands']]

        if operator == '+':
            return sum(operands)
        elif operator == '-':
            if len(operands) == 1:
                return -operands[0]  # Handle unary minus
            else:
                return operands[0] - sum(operands[1:])
        elif operator == '*':
            result = 1
            for operand in operands:
                result *= operand
            return result
        elif operator == '/':
            if len(operands) == 0:
                raise ValueError("Division requires at least one operand")
            result = operands[0]
            for operand in operands[1:]:
                if operand == 0:
                    raise ZeroDivisionError("Division by zero")
                result /= operand
            return result
        else:
            raise ValueError("Unsupported operator: " + operator)
    else:
        raise ValueError("Invalid expression format")

# Example usage:
expression1 = {'operator': '+', 'operands': [{'value': 5}, {'value': 3}]}
print(f"Expression 1: {expression1}")
print(f"Result: {evaluate_expression(expression1)}")

expression2 = {'operator': '*', 'operands': [{'value': 2}, {'operator': '+', 'operands': [{'value': 4}, {'value': 1}]}]}
print(f"\nExpression 2: {expression2}")
print(f"Result: {evaluate_expression(expression2)}")

variables = {'x': 10, 'y': 2}
expression3 = {'operator': '-', 'operands': [{'variable': 'x'}, {'operator': '/', 'operands': [{'variable': 'y'}, {'value': 2}]}]}
print(f"\nExpression 3: {expression3}")
print(f"Result: {evaluate_expression(expression3, variables)}")
```
