2. Write a program to display the cos(x) value where x ranges from 0 to 360 in steps of 15.


```python
import math

print("Angle (Degrees)\tcos(x)")

for x in range(0, 361, 15):
    rad = math.radians(x)   # Convert degrees to radians
    result = math.cos(rad)
    print(x, "\t", round(result, 4))

```

3. Write a Python program that: Accepts an odd integer n
- Prints a diamond rotated by 45 degrees
- Diamond must be centered
- Use coordinate-based logic only

```python
n = int(input("Enter number of rows: "))

for i in range(n):
    # print leading spaces
    print("  " * (n - i - 1), end="")
    # print stars
    print("*  " * n)
print()
```
