Q1. Generate all prime numbers less than 100.Print them in a spiral format (like a clockwise
square spiral) of minimal size that fits all primes.

```python
for num in range(2, 100):
    for i in range(2, num):
        if (num % i) == 0:
            break
    else:
        print(num)
```

# Task
Generate a square spiral of prime numbers less than 100.

## Generate prime numbers

### Subtask:
Get the list of prime numbers less than 100.


**Reasoning**:
Initialize an empty list and iterate through numbers to find primes, appending them to the list.



```python
prime_numbers = []
for num in range(2, 100):
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if (num % i) == 0:
            is_prime = False
            break
    if is_prime:
        prime_numbers.append(num)

print(prime_numbers)
```

## Determine spiral dimensions

### Subtask:
Calculate the minimal size of the square spiral that can fit all prime numbers.


**Reasoning**:
Get the total number of prime numbers and calculate the smallest integer n such that n*n is greater than or equal to the total number of prime numbers to determine the minimal size of the square spiral matrix.



```python
import math

num_primes = len(prime_numbers)
spiral_size = math.ceil(math.sqrt(num_primes))

print(f"Total number of prime numbers: {num_primes}")
print(f"Minimal size of the square spiral: {spiral_size}x{spiral_size}")
```

## Initialize spiral matrix

### Subtask:
Create a 2D list (matrix) with the determined dimensions, initialized with zeros or a placeholder.


**Reasoning**:
Create a 2D list (matrix) with the determined dimensions, initialized with zeros or a placeholder.



```python
spiral_matrix = [[0 for _ in range(spiral_size)] for _ in range(spiral_size)]
for row in spiral_matrix:
    print(row)
```

## Fill spiral matrix

### Subtask:
Populate the matrix with the prime numbers in a clockwise square spiral pattern.


**Reasoning**:
Populate the initialized matrix with the prime numbers in a clockwise square spiral pattern by iterating through the prime numbers, placing them in the current position, and updating the position and direction based on the spiral boundaries.



```python
row, col = 0, 0
direction = 0  # 0: right, 1: down, 2: left, 3: up
top, bottom, left, right = 0, spiral_size - 1, 0, spiral_size - 1

for prime in prime_numbers:
    spiral_matrix[row][col] = prime

    if direction == 0:  # right
        if col < right:
            col += 1
        else:
            direction = 1
            top += 1
            row += 1
    elif direction == 1:  # down
        if row < bottom:
            row += 1
        else:
            direction = 2
            right -= 1
            col -= 1
    elif direction == 2:  # left
        if col > left:
            col -= 1
        else:
            direction = 3
            bottom -= 1
            row -= 1
    elif direction == 3:  # up
        if row > top:
            row -= 1
        else:
            direction = 0
            left += 1
            col += 1

for row in spiral_matrix:
    print(row)
```

## Print spiral

### Subtask:
Display the spiral matrix in a readable format.


**Reasoning**:
Iterate through the spiral matrix and print each element followed by a space, then print a newline after each row to display the spiral in a readable format.



```python
for row in spiral_matrix:
    for number in row:
        print(number, end=" ")
    print()
```

Q2. You are a data analyst at a streaming platform, managing user interactions, analytics,
and daily operations. Use a for loop to create a text-based graph of hourly views. For
example, use * to represent every 5 views.

```python
import random

# Simulate hourly views for a day
hourly_views = [random.randint(10, 100) for _ in range(24)]

print("Hourly Views Graph:")
print("-" * 40)

# Create the text-based graph
for hour in range(24):
    views = hourly_views[hour]
    asterisks = '*' * (views // 5)
    print(f"Hour {hour:02d}: {asterisks} ({views} views)")

print("-" * 40)
```

Q3. Each video watched generates revenue based on ad impressions:
* First 10 views: $0.50 per view
*  Next 20 views: $0.30 per view
*  Remaining views: $0.10 per view

You are given hourly views for the day. Calculate total daily revenue.

```python
# Calculate total daily revenue
total_daily_revenue = 0

for views in hourly_views:
    revenue_for_hour = 0
    remaining_views = views

    # First 10 views
    if remaining_views > 0:
        first_10_views = min(remaining_views, 10)
        revenue_for_hour += first_10_views * 0.50
        remaining_views -= first_10_views

    # Next 20 views
    if remaining_views > 0:
        next_20_views = min(remaining_views, 20)
        revenue_for_hour += next_20_views * 0.30
        remaining_views -= next_20_views

    # Remaining views
    if remaining_views > 0:
        revenue_for_hour += remaining_views * 0.10

    total_daily_revenue += revenue_for_hour

print(f"\nTotal daily revenue: ${total_daily_revenue:.2f}")
```
